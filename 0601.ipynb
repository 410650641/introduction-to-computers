```python
k=int(input("請輸入正整數："))

for i in range(0,k):
  for j in range(0,k-i):
    print('*',end="")
  print("\n",end="")
b=int(input('請輸入正整數：'))
for i in range(0,b):
  for j in range(0,b-i):
    print('*',end="")
  print("\n",end="")
```
