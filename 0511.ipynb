```python
temp = float(input('請輸入您的體溫'))
if temp >=37.5:
  print('體溫過高！')
elif temp >=37 and temp <37.4:
  print('體溫略高！')
elif temp >=34 and temp <36.9:
  print('體溫正常！')
else:
  print('體溫異常，請重新量測！')

```

```python
money = int(input('請輸入購物金額：'))
if(money>=10000):
  if(money>=100000):
    print('八折',money*0.8,end="元\n")
  elif(money>=30000):
    print('九折',money*0.9,end="元\n")
  elif(money>=50000):
    print('八五折',money*0.85,end="元\n")
  else:
    print('九五折',money*0.95,end="元\n")
    
else:
  print('沒打折',money,end="元\n")
```

```python
month=int(input('請輸入月份：'))
if month>=3 and month<=5:
  print(month,'月是春天！')
elif month>=6 and month<=8:
  print(month,'月是夏天！')
elif month>=9 and month<=11:
  print(month,'月是秋天！')
elif month>=12 and month<13 and month >0 and month<=2:
  print(month,'月是冬天！')
else:
  print('月份不在範圍內！')
```

```python
money = int(input('請輸入今年收入淨額：'))
if(money>=300000):
  if(money>=600000):
    print(money*0.13,end="元\n")
  elif(money>=1000000):
    print(money*0.21,end="元\n")
  elif(money>=2000000):
    print(money*0.3,end="元\n")
  else:
    print(money*0.06,end="元\n")
    
else:
  print('無稅率',money,end="元\n")
```
